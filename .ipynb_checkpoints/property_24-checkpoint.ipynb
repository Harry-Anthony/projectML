```python
df = df.drop(['Bedroom 1','Bedroom 2', 'Bedroom 3', 'Bathroom 1', 'Bathroom 2', 'Listing Date', 'Listing Number', 'Price per mÂ²', 'Description', 'Lounges', 'Dining Rooms', 'Lifestyle', 'Domestic Rooms', 'Entrance Hall', 'Reception Rooms', 'Dining Room', 'Lounge', 'Nearby Public Transport', 'Street Address', 'Levies', 'Kitchens', 'Kitchen'], axis = 1)
```

```python
df = df.drop(['Security'], axis = 1)
df['Erf Size'] = df['Erf Size'].str.replace(' mÂ²','')
df['price'] = df['price'].str.replace('R ', '')
df['Rates and Taxes'] = df['Rates and Taxes'].str.replace('R ', '')
df['Floor Size'] = df['Floor Size'].str.replace(' mÂ²', '')
```

```python
df
```

```python
df['Erf Size'].str.contains("ha").value_counts()
```

```python
dff = df.copy()
```

```python
def replace_ha(n):
    if (n.str.contains('ha')):
        n.str.replace(' ha', '')
        return n * 10000
    else:
        return n
```

```python
dff['Erf Size'] = dff['Erf Size'].map(replace_ha)
```
